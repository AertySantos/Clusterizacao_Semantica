```python
# Lista para armazenar colunas a serem removidas
drop_columns = []
df2 = pd.DataFrame()  # Criar um DataFrame vazio para armazenar as novas colunas

for col in df.columns:
    if not pd.api.types.is_numeric_dtype(df[col]):  # Verifica se a coluna é numérica
        print(f"Coluna não numérica: {col}")
        # Criar colunas dummy com um prefixo
        dummies = pd.get_dummies(df[col], prefix=col)
        # Concatenar as colunas dummy ao DataFrame auxiliar
        df2 = pd.concat([df2, dummies], axis=1)
        # Armazenar a coluna original para remoção posterior
        drop_columns.append(col)

# Remover as colunas originais após a iteração
df = df.drop(columns=drop_columns)

# Concatenar df2 ao df original corretamente
df = pd.concat([df, df2], axis=1)

# Opcional: resetar o índice se necessário
df.reset_index(drop=True, inplace=True)


```

```python
# Converte colunas booleanas para inteiros
#bool_cols = df.select_dtypes(include=['bool']).columns
#df[bool_cols] = df[bool_cols].astype(int)


# Identifica o nome da primeira coluna
col0 = df.columns[0]

# Normaliza somente a coluna 0
df[col0] = (df[col0] - df[col0].min()) / (df[col0].max() - df[col0].min())
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Empilhando as colunas para criar um único array 2D
array_2d = np.column_stack([df[col].values for col in df.columns])

# Aplica PCA para reduzir para 2 dimensões
pca = PCA(n_components=2, svd_solver='full')
X_transformed = pca.fit_transform(array_2d)

# Adiciona os resultados do PCA ao DataFrame
df['pca_1'] = X_transformed[:, 0]
df['pca_2'] = X_transformed[:, 1]

# Salvar as colunas pca_1 e pca_2 sem cabeçalhos
df[['pca_1', 'pca_2']].to_csv("data/input/dataset.txt", sep=" ", index=False, header=False)

# Exibe o DataFrame com os resultados
#print(df)
```
